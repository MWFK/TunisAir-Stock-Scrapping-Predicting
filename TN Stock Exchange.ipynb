```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains

# 'BS' doe not exist
TN_Listed_Stock_Companies = ['AB', 'ADWYA', 'AETEC', 'AL', 'AMS', 'ARTES', 'ASSAD', 'ATB', 'ATL', 'BH', 'BIAT', 'BNA', 'BT', 'CC',
                             'CELL', 'CITY', 'CREAL', 'DH', 'ECYCL', 'GIF', 'ICF', 'LNDOR', 'LSTR', 'MGR', 'MNP', 'MPBS', 'NAKL', 'NBL',
                             'OTH', 'PLAST', 'POULA', 'SAH', 'SAMAA', 'SERVI', 'SFBT', 'SIAM', 'SMG', 'SOKNA', 'SOMOC', 'SOPAT','SOTE',
                             'STAR', 'STB', 'STPAP', 'STPIL', 'STVR', 'TAIR', 'TGH', 'TLNET', 'TLS', 'TPR', 'TRE', 'UADH', 'UIB','UMED'] 

for i in range(len(TN_Listed_Stock_Companies)):
    
    driver = webdriver.Chrome()
    
    Link = "https://www.ilboursa.com/marches/download.aspx?s=" +  str(TN_Listed_Stock_Companies[i])
    driver.get(Link)
    
    sleep(3)

    datefield_from = driver.find_element_by_id('ctl00_BodyABC_txtFrom')
    datefield_to   = driver.find_element_by_id('ctl00_BodyABC_txtTo')

    month_from = -2
    month_to   = 0
    
    sleep(3)
    
    if driver.find_element_by_id('hideModalPopupViaClientButton').is_displayed() : #if driver.find_element_by_id('hideModalPopupViaClientButton').size['width'] != 0 # if len(driver.find_element_by_id('hideModalPopupViaClientButton'))
            driver.find_element_by_id('hideModalPopupViaClientButton').click()
            #driver.quit()
            #driver = webdriver.Chrome()
            #driver.get(Link)
        
    for i in range(4):
        
        if driver.find_element_by_id('hideModalPopupViaClientButton').is_displayed() : #if driver.find_element_by_id('hideModalPopupViaClientButton').size['width'] != 0 # if len(driver.find_element_by_id('hideModalPopupViaClientButton'))
            driver.find_element_by_id('hideModalPopupViaClientButton').click()
            #driver.quit()
            #driver.refresh();
            #driver = webdriver.Chrome()
            #driver.get(Link)

        month_from += 3
        month_to   += 3

        date_from = "01/" + str(month_from) + "/2020"
        date_to   = "28/" + str(month_to)   + "/2020"
        
        # https://stackoverflow.com/questions/18225997/stale-element-reference-element-is-not-attached-to-the-page-document
        ActionChains(driver).move_to_element(datefield_from).click().click().click().send_keys(date_from).perform()
        sleep(1)
        ActionChains(driver).move_to_element(datefield_to).click().click().click().send_keys(date_to).perform()
        sleep(1)
        ActionChains(driver).move_to_element(datefield_from).click().click().click().send_keys(date_from).perform()
        sleep(1)
        ActionChains(driver).move_to_element(datefield_to).click().click().click().send_keys(date_to).perform()
        sleep(1)
        driver.find_element_by_id("ctl00_BodyABC_Button1").click()
        
        sleep(2)
    sleep(2)

    driver.quit()
```
